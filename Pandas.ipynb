# Remoção de dados

```
  .drop([column(s)], axis=0/1, inplace=True/False)

  axis: 0 para as linhas e 1 para as colunas

  inplace: Se True, altera o DataFrame
```

````python
corona.head(1)
````

````python
# Excluindo uma coluna
corona.drop(['Province/State'], axis=1, inplace=True)
corona.head(1)
````

````python
df = corona.drop(['Confirmed'], axis=1)
df.head(1)
````

````python
corona.head(1)
````

# Aplicação de Filtros

````python
# Retornando uma Series da variável Country/Region
corona['Country/Region']
````

````python
# Retornando um DataFrame da variável Country/Region
corona[['Country/Region']]
````

````python
# Retornar mais de uma coluna
corona[ ['Country/Region', 'Recovered'] ]
````

````python
# Vamos ver um comando mais personalizável
corona.query('Recovered > 50000')
````

````python
# Filtragem de variáveis qualitativas
corona[corona['Country/Region'] == 'Brazil']
````

# Como salvar filtros?

````python
# Aplicamos um filtro somente para o país Brasil.
corona_brasil = corona[corona['Country/Region'] == 'Brazil']
corona_brasil.shape
````

````python
# Vamos salvar o resultado em um novo "dataset" (arquivo de dados)
corona_brasil.to_csv('/content/corona_brasil.csv', index=None)
````
