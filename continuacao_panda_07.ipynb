# Criação de subsets
São sub-conjuntos de um DataFrame, os quais `não alteram` o valor do DataFrame

```python
colunas = ['Country/Region', 'Confirmed', 'Deaths']
df[colunas].sample(5)
```

```python
# Selecione três conjuntos de colunas e salve os resultados em arquivos ".CSV"
```

```python
# Criando novas colunas a partir do resultado das existentes
df['Confirmed + Deaths'] = df['Confirmed'] + df['Deaths']
df.sample(5)
```

```python
# Criar uma nova coluna que mostre a relação entre mortes e casos confirmados.
df['Death by Cases'] = df['Deaths'] / df['Confirmed']
df.sample(5)
```

# Conversão de Valores
* `.to_numeric()`
* `.to_datetime()`
* `.astype()`
    - `'bool'`
    - `'int64'`
    - `'float64'`
    - `'str'`

```python
# verificando novamente o DataFrame
df.info()
```

```python
df.head()
```

```python
# alterando os valores do DataFrame (values)
df['Country/Region'].replace('Mainland China', 'China', inplace=True)
df.sample(10)
```

```python
df.head()
```

```python
df['Country/Region'].replace('US', 'United States', inplace=True)
df.tail(10)
```

```python
df.head()
```

```python
# Filtrando pelo país
df[df['Country/Region'] == 'Brazil']
```

```python
# Antes de preencher os NaN
df.head()
```

```python
# Com base nos dados, percebemos que há um NaN, ou seja, um valor ausente.
# Precisamos urgentemente corrigir isso!!!!!!
# Para isso, vamos utilizar o .fillna()
df.fillna(0, axis=1, inplace=True)
df.head()
```

```python
# Alterar o nome de colunas
df.rename(    
    columns={
        'Country/Region' : 'Country'
    },
    inplace=True
    )
df.head()
```

```python
# Vamos corrigir o tipo de dados dos campos 'data'
df.info()
```

```python
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])
df.info()
```

Remoção de Dados Duplicados
* `.unique()`: retorna valores únicos
* `.duplicated()`: retorna `bool` para linhas duplicadas
* `drop_duplicated()`: retorna *DataFrame* com linhas duplicadas removidas

```python
# Quantos países existem nos dados?
print('Quantidade de países:', len(df['Country'].unique()))
```

```python
# O que acontece se não usarmos o len()?
df['Country'].unique()
```

```python
# Quantos países estão duplicados?
98251-223
```

```python
# Outra forma
sum(df.duplicated(subset=['Country']))
```

```python
# Removendo os duplicados
# keep='last' mantém uma cópia de cada registro
df.drop_duplicates(subset=['Country'], keep='last', inplace=True)
df.info()
```

# Ordenação e Ranqueamento
* `.sort_index()`
* `.sort_values()`: ordena de acordo com uma coluna
* `.rank()`: computa ranks numéricos ($1$ à $n$)

```python
# Ordenação crescente
# ascending=True -> crescente
df.sort_values(by=['Recovered'], ascending=True).head()
```

```python
# Ordenação decrescente
# ascending=False -> decrescente
df.sort_values(by=['Recovered'], ascending=False).head()
```

```python
df.sort_values(by=['Country'], ascending=False).head()
```
